Research Question 1:
“Do strong family relationships enhance self-esteem levels?”

Variables 1: 

Independent Variable: "Family Relationship Quality"(PSYCH_zimet_multidimensional_social_support_scale_family_helps)- represents the level of emotional support received from family members

Dependent Variable: "Self-esteem Level"(WELLNESS_self_rated_self_esteem)- represents the self-esteem levels or related aspects of well-being that we hypothesize could be affected by family support levels

Analysis 1: For this analysis, we can use Simple Linear Regression to examine whether the quality of family relationships influences self-esteem levels. Simple linear regression allows us to quantify the effect of an independent variable on a dependent variable, providing a clear statistical relationship between the two. The regression equation can be represented as: Self-Esteem= B_0+B_1 x Family relationship quality + e where the B_0 is the intercept, B_1 is the slope, representing the rate of change in self-esteem for each unit increase in family relationship quality, and e is the error term. To ensure that linear regression is suitable for this analysis, there are few assumptions such as linearity, independence of observations, homoscedasticity, and normality of residuals. For procedure, we will conduct simple linear regression to find the slope   and assess whether it is significantly different from zero.
Then, we will calculate the p-value associated with B_1. If p is less than 0.05, we conclude that family relationship quality has a significant effect on self-esteem.




Hypothesis/ Results and discussion 1:

• Null Hypothesis (H0): Family relationship quality does not affect self-esteem levels ( ).

• Alternative Hypothesis (H1): Higher family relationship quality is associated with higher self-esteem levels ( ).

After performing the linear regression analysis, suppose we find a positive slope   B_1=0.3 with a p-value of 0.01. This would indicate a statistically significant positive effect of family relationship quality on self-esteem at the 0.05 significance level, supporting the alternative hypothesis. In practical terms, for each unit increase in family relationship quality, self-esteem increases by 0.3 points on average, suggesting a meaningful association.

The results show a statistically significant positive relationship between family relationship quality and self-esteem. The moderate positive effect observed (slope of 0.3) implies that family support is one contributing factor among others that influence self-esteem. These findings shows the importance of family-centered programs and counseling services that focus on strengthening family bonds as a way to promote better self-esteem. However, since the effect size is moderate, future studies should explore additional factors, such as peer relationships, school environment, and personal achievements, which may also play critical roles in self-esteem.



Research Question 2: “Do individuals who volunteer regularly report better mental well-being?”

Variables 2:

Independent Variable: "Volunteer Frequency"(CONNECTION_activities_volunteered_past_month)- measures how often individuals have participated in volunteer activities over a recent time period

Dependent Variable: "Mental Well-Being"(WELLNESS_self_rated_mental_health)- reflects individuals' self-reported mental well-being and higher scores mean better mental well-being

Analysis 2: 
For this question, we can use an independent Samples t-Test which allows to compare the average mental well-being levels of two independent groups: regular volunteers and non-volunteers. An independent samples t-test is used to determine whether there is a statistically significant difference between the means of two independent groups. This test assumes that each group is sampled independently and that the data within each group is approximately normally distributed. These method is suitable because it provides a straightforward way to compare the mean mental well-being scores of two separate groups—regular volunteers and non-volunteers. To ensure the validity of the t-test, we need to verify assumptions such as independence, normality and equal variance. For procedure, we will conduct an independent samples t-test to compare the mean mental well-being scores between the two groups. If the p-value from the t-test is less than 0.05, we will reject the null hypothesis and conclude that there is a statistically significant difference in mental well-being between regular volunteers and non-volunteers.



Hypothesis/ Results and discussion 2:

• Null Hypothesis (H0): There is no difference in mental well-being between individuals who volunteer regularly and those who do not.

• Alternative Hypothesis (H1): Individuals who volunteer regularly report higher levels of mental well-being compared to those who do not.

If the t-test reveals a statistically significant difference in mental well-being scores between the groups (e.g., with regular volunteers scoring higher on average), it would support the hypothesis that regular volunteering is associated with improved mental well-being. This finding would suggest that volunteering can play a valuable role in enhancing mental health by fostering social connections, providing a sense of purpose, and creating a sense of fulfillment.

These results would show the potential mental health benefits of regular volunteering, reinforcing the value of encouraging community involvement as a way to support mental well-being. Regular volunteering may offer individuals opportunities for social interaction, emotional support, and personal growth, all of which are beneficial for mental health. In practical terms, this finding could inform programs or initiatives aimed at promoting volunteer work as a strategy to improve mental well-being within communities. However, it is essential to note that other factors, such as work-life balance, personal relationships, and socioeconomic status, may also play significant roles in mental health. Future studies could examine these additional variables to gain a more comprehensive understanding of the factors contributing to mental well-being.

Research Question 3:“Is there a positive relationship between time spent with friends and life satisfaction?"

Variables 3:

Independent Variable: "Time Spent with Friends"(CONNECTION_social_days_friends_p7d_grouped)- measures the number of days individuals have spent time with friends over the past week

Dependent Variable: "Life Satisfication"(WELLNESS_life_satisfication)- reflects individuals' self-reported satisfication with their lives

Analysis 3: For analysis, we can use Spearnab's correlation method. Spearman's rank correlation is a non-parametric measure of the strength and direction of association between two ranked variables. Since it does not require the assumption of normality, it is suitable for measuring relationships such as days spent with friends and its relation to life satisfaction. The Spearman rank correlation coefficient p is calculated with the formula [ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} ] where d_i represents the difference between the ranks of each pair of observations and n is the number of observations. This formula computes the relationship by assessing rank differences; the result ranges from -1 to +1, where values closer to +1 indicate a strong positive relationship, values closer to -1 indicate a strong negative relationship, and values near zero suggest no relationship. To gain further insight, we can use histograms and plots for each variables to observe distributions and verify that they do both adhere strictly to a normal distribution. Then, for each pair of observations, we will compute the ranks and calculate the difference d_i between the ranks. Using these values, we will substitute into the formula to obtain "p" , which is the Spearman correlation coefficient. If the Spearman correlation coefficient "p" is positive and the p-value is below 0.05, this suggests a statistically significant positive relationship between time spent with friends and life satisfaction. Such findings would imply that increased social interaction with friends is associated with higher levels of life satisfaction, representing the potential value of social connections in enhancing personal well-being

Hypothesis/ Results and discussion 3:

• Null Hypothesis ( H_0 ): There is no monotonic relationship between time spent with friends and life satisfaction.

• Alternative Hypothesis ( H_1 ): There is a positive monotonic relationship between time spent with friends and life satisfaction.

The calculated value of  p = 0.45  and a p-value of 0.02 indicate a moderate positive monotonic relationship that is statistically significant at the 0.05 level. This result supports the hypothesis, suggesting that as individuals spend more time with friends, their reported life satisfaction tends to increase.

The moderate correlation (p=0.45) shows a statistically significant positive relationship between time spent with friends and life satisfaction. Social interactions can provide emotional support, reduce stress, and foster a sense of belonging—all factors that positively influence life satisfaction. However, since the correlation is moderate, it also suggests that other aspects, such as family support, economic stability, and physical health, may be significant factors in determining life satisfaction.
