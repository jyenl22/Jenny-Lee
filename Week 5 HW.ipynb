8. Fisher's Tea Experiment Data Analysis Assignment

Introduction:  
Fisher's Tea Experiment is an example of hypothesis testing where the goal is to determine if an observed pattern (correctly identifying tea pouring order) can be attributed to chance or not. In this assignment, the context is extended to a group of students instead of an individual. Out of a population of 80 students, 49 students were randomly sampled to determine whether they could correctly identify the pouring order of tea and milk. This analysis will test whether the observed success rate of 49 correct guesses is consistent with random chance or suggests a greater-than-chance ability.

Hypothesis:  

- Null Hypothesis: The students guess the pouring order of tea and milk purely by chance. Specifically, the probability of a correct guess is \(p = 0.5\).
  
- Alternative Hypothesis: The students have an ability to identify the pouring order of tea and milk that exceeds random guessing. Specifically, the probability of a correct guess is \(p > 0.5\).
  

Null Hypothesis assumes that any observed success rate is due to random guessing and alternative hypothesis suggests that the observed success rate indicates a meaningful ability beyond chance.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data
total_students = 80
correct_guesses = 49
p_null = 0.5  # Null hypothesis: p = 0.5 (random guessing)

# 1. Observed proportion of correct guesses
observed_proportion = correct_guesses / total_students

# 2. Standard error of the proportion under H₀
se_null = np.sqrt(p_null * (1 - p_null) / total_students)

# 3. Z-test statistic
z_score = (observed_proportion - p_null) / se_null

# 4. p-value for one-sided test (since we are testing p > 0.5)
p_value = 1 - norm.cdf(z_score)

# 5. Simulation under H₀: Generate 10000 random samples of correct guesses under the null hypothesis
np.random.seed(42)  # For reproducibility
simulated_data = np.random.binomial(n=total_students, p=p_null, size=10000)
simulated_proportions = simulated_data / total_students

# 6. Plot the distribution of simulated proportions and observed proportion
plt.figure(figsize=(10, 6))
plt.hist(simulated_proportions, bins=50, density=True, alpha=0.7, color='skyblue')
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f"Observed Proportion: {observed_proportion:.2f}")
plt.title("Simulated Proportions of Correct Guesses Under H₀")
plt.xlabel("Proportion of Correct Guesses")
plt.ylabel("Density")
plt.legend()
plt.show()

# Output results
print(f"Observed Proportion: {observed_proportion:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
```

Data Analysis:

The observed proportion of correct guesses is `49 / 80 = 0.6125`. Under the null hypothesis (H₀), the standard error for a proportion is computed using the formula:

$$
SE = \sqrt{\frac{p (1 - p)}{n}}
$$

where \( p = 0.5 \) and \( n = 80 \). Using this standard error, the z-score is calculated to measure the distance between the observed proportion and the expected proportion under H₀, expressed in standard error units. For the given data, the z-score is approximately 2.02. 

This z-score corresponds to a p-value of 0.0221 for a one-sided test. A p-value of 0.0221 indicates that there is a 2.21% probability of observing a proportion as extreme as 0.6125 if the null hypothesis is true. Since the p-value is less than the common significance level of 0.05, H₀ is rejected, suggesting that students’ performance is unlikely to be due to random guessing.


A simulation was conducted under the null hypothesis to supplement the analytical results. Using a binomial distribution, 10,000 samples of 80 students were generated, each with a success probability of \( p = 0.5 \). For each sample, the proportion of correct guesses was calculated, and the resulting distribution was visualized using a histogram. The histogram demonstrates the variability of proportions expected under \( H_0 \), with the observed proportion of 0.6125 marked by a red dashed line. This visualization shows that the observed proportion lies in the upper tail of the null distribution, supporting the rejection of \( H_0 \). Overall, the statistical analysis and simulation results indicate that the observed success rate of 49 correct guesses out of 80 students is significantly greater than what would be expected by random chance. With a p-value of 0.0221, there is sufficient evidence to reject the null hypothesis at the 0.05 significance level, suggesting that the students’ ability to identify the pouring order exceeds what would be expected from random guessing.
