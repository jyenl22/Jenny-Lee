## 1. Confusion Matrix and Basic Metrics Calculation

The confusion matrix data is as follows:
- True Positive (TP): 23
- False Negative (FN): 7
- False Positive (FP): 90
- True Negative (TN): 280

### (1) Accuracy Calculation
Formula: (TP + TN) / (TP + FN + FP + TN)

Calculation: (23 + 280) / (23 + 7 + 90 + 280) = 303 / 400 = 0.759

Result: Accuracy = 0.759

### (2) Sensitivity Calculation
Formula: TP / (TP + FN)

Calculation: 23 / (23 + 7) = 23 / 30 = 0.767

Result: Sensitivity = 0.767

### (3) Specificity Calculation
Formula: TN / (FP + TN)

Calculation: 280 / (90 + 280) = 280 / 370 = 0.757

Result: Specificity = 0.757

## 2. Additional Metrics Calculation

### (1) False Positive Rate (FPR) Calculation
Formula: 1 - Specificity

Calculation: 1 - 0.757 = 0.243

Result: FPR = 0.243

### (2) Precision Calculation
Formula: TP / (TP + FP)

Calculation: 23 / (23 + 90) = 23 / 113 ≈ 0.204

Result: Precision ≈ 0.204
